Ans. Data cleaning (also called data cleansing or data preprocessing )is the process of identifying and correcting (removeing) errors, inconsistencies,and inaccuracies in a dataset to make it accurate ,complete ,and ready for analysis.
**WHY DATA CLEANING IS IMPORTANT IN DATA ANALYS**
1. Improve accuracy
2. Save time
3. Better decision making
4. Prevents model performance
**>POTENTIONAL CONSEQUENCES OF THE ANALYZING MESSY DATA:**
1. Inaccurate results.
2. Misleading insight.
3. Poor decision making
4. Loss of trust
5. Wasted time and resources .
6. Faulty model and predictions.

1. Data collection and importing .
2. Data inspection
3. Handling missing data .
4. Removing duplicates
5. Correcting errors and inconsistencies
6. Handling outliners
7. Data transformation
8. Data organization and formatting
