SOLUTION:- TODAY(): RETURN 2024-09-16(assuming today's data is September16,2024)

USE CASES:-
* USE TODAY() WHEN:-
> Calculating age or tenure
>Creating data based filters
*Use NOW() when:-
> Tracking real -time data (eg. logging events)
>Creating time -sensitive metric
> Creating timestamped records.
