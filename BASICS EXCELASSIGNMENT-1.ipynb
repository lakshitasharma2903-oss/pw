Que.1 Explain the difference between Relative,Absolute,Mixed cell referencing.

Ans. The difference between Relative,Absolute,Mixedcell referencing

# Relative cell referencing

✒ The reference changes when you copy the formula to another cell.

✒  Example:

In cell C1,you write =A1+B1.

If you copy it to C2, Excel automatically changed it to =A2+B2.

#Absolute Reference

✒ The reference does not change when you copy the formula.

✒ Example

In cell C1,you write =A1*$B$1.

If you copy this down to C3,it stays =A2*$B$1.

So $B$1 is fixed- it always refers to that one cell.

# Mixed Cell Reference

✒ Only part of the reference changes:

*$A1:Column Astays the same row changes.

*A$1:Row1 stays the same ,column changes.

*Example:Used in the tables where you want to keep one direction (like  column or row ) fixed.
